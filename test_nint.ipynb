```python
import numpy as np
```

```python
def f(x):
    return np.exp(x)
```

```python
x0 = 0
x1 = np.pi/2
```

```python
def intg(Nmesh):
    dxar = np.linspace(x0,x1, Nmesh)
    far = [f(x) for x in dxar]
    intg = (x1-x0)*np.sum(far)/Nmesh
    return intg
```

```python
Nmesh = np.arange(1,100,1)
intar = []
for Nm in Nmesh:
    intar.append(intg(Nm))
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.plot(Nmesh, intar, marker='s')   
```
